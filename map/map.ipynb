```python
import pandas as pd

hospitals = pd.read_json('../data/merged_file_junrang_index_db.json')
hospitals.head(2)
```

```python
hospitals.info()
```

```python
# object 타입 -> numeric으로 변환
hospitals['latitude'] = hospitals['latitude'].astype(float)
hospitals['longitude'] = hospitals['longitude'].astype(float)
hospitals.info()
```

```python
# 변환안되는 널값은 바꿔줌
hospitals['latitude'] = pd.to_numeric(hospitals['latitude'], errors='coerce')
hospitals['longitude'] = pd.to_numeric(hospitals['longitude'], errors='coerce')
```

```python
hospitals[hospitals['latitude'].isnull()]
```

```python
from geopy.distance import geodesic # geopy에서의 위도와 경도는 (latitude, longitude) 순서로 사용

# 중랑구 구민회관
user_latitude = 37.573872
user_longitude = 127.080849

# 사용자의 위치
user_location = (user_latitude, user_longitude)

# 사용자가 정한 반경 (단위: 킬로미터)
radius = 5

# 반경 내에 있는 병원 필터링
nearby_hospitals = []

# 중복된 병원 이름을 처리하여 유일한 병원만을 남기는 작업
unique_hospitals = {}
for _, hospital in hospitals.iterrows():
    name = hospital["name"]
    location = (hospital["latitude"], hospital["longitude"])
    distance = geodesic(user_location, location).kilometers

    if distance <= radius and (name not in unique_hospitals or distance < unique_hospitals[name]["distance"]):
        unique_hospitals[name] = {"name": name, "distance": distance, "latitude": hospital["latitude"], "longitude": hospital["longitude"]}

# 결과 출력
for hospital in nearby_hospitals:
    print(f"{hospital['name']} - Latitude: {hospital['latitude']}, Longitude: {hospital['longitude']}")
```

```python
from haversine import haversine

# 중랑구 구민회관
user_latitude = 127
user_longitude = 37


# 사용자의 위치
user_location = (user_latitude, user_longitude)

# 디비에서 가져온 병원 리스트 (예시)
hospitals

# 사용자가 정한 반경 (단위: 킬로미터)
radius = 5

# 반경 내에 있는 병원 필터링
nearby_hospitals = []

for _, hospital in hospitals.iterrows():
    hospital_location = (hospital["lat"], hospital["lng"])
    hospital_coords = (float(hospital["lat"]), float(hospital["lng"]))
    print(f"Hospital: {hospital['name']} - Latitude: {hospital_coords[0]}, Longitude: {hospital_coords[1]}")
    distance = haversine(user_location, hospital_coords, unit='km')

    if distance <= radius:
        nearby_hospitals.append({
            "name": hospital["name"],
            "latitude": hospital_coords[0],
            "longitude": hospital_coords[1]
        })


# 결과 출력
for hospital in nearby_hospitals:
    print(f"{hospital['name']} - 위도: {hospital['latitude']}, 경도: {hospital['longitude']}")
```
