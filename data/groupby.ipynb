```python
import pandas as pd
```

```python
df = pd.read_excel('../data/review_file_junrang.xlsx')
df.head(2)
```

```python
# 'name'별로 'sentiment'가 0인 경우와 1인 경우의 개수 계산
sentiment_counts = df.groupby(['name', 'sentiment']).size().unstack(fill_value=0)
sentiment_counts
```

```python
# 'name'별로 'sentiment'와 'content' 개수 계산
result = df.groupby('name').agg(부정=('sentiment', lambda x: (x == 0).sum()),
                                긍정=('sentiment', lambda x: (x == 1).sum()),
                                review_total=('content', 'count')) # nunique는 중복값 제외됨
result
```

```python
result.to_excel('groupby_junrang.xlsx')
```
