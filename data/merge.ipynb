```python
import pandas as pd
```

```python
df1 = pd.read_excel('C:/Users/user/Desktop/saltlux_project/data/1.병원정보서비스.xlsx')
df1.head(2)
```

```python
# 필요한 컬럼 지정 : 요양기관명, 종별코드명, 우편번호, 주소, 전화번호, 좌표(X), 좌표(Y)

df1 = df1[['요양기관명', '종별코드명', '우편번호', '주소', '전화번호', '좌표(X)', '좌표(Y)']]
df1.sample(2)
```

```python
# 서울지역 데이터만 사용 -> 주소 컬럼에서 서울특별시로 시작하는 데이터만 선택

seoul_data = df1[df1['주소'].str.startswith('서울특별시')]
seoul_data.sample(2)
```

```python
df2 = pd.read_excel('C:/Users/user/Desktop/saltlux_project/data/5.의료기관별상세정보서비스_03_진료과목정보.xlsx')
df2.head(2)
```

```python
# 필요한 컬럼 지정 : 요양기관명, 진료과목

df2 = df2[['요양기관명', '진료과목코드명']]
df2.sample(2)
```

```python
# seoul_data에 있는 요양기관만 조인

merged_df = pd.merge(seoul_data, df2, on='요양기관명', how='inner')
merged_df.sample(2)
```

```python
# 엑셀 파일로 저장
merged_df.to_excel('merged_file_seoul.xlsx', index=False)
```
